```python
import numpy as np
from sklearn import datasets
```

<div style="background:#1F3A68;color:#FFFFFF;padding:12px 18px;border-radius:10px;margin:10px 0 14px 0;display:flex;align-items:center;gap:14px;box-shadow:0 2px 6px rgba(0,0,0,0.15);"><div style="background:rgba(255,255,255,0.22);color:#FFFFFF;font-weight:800;padding:6px 16px;border-radius:20px;font-size:22px;">2</div><div style="font-size:30px;font-weight:700;color:#FFFFFF;">Load the dataset (from scikit-learn)</div></div>

<div style="font-size:20px;line-height:1.75;margin:6px 0;">Load a 2-feature <b>make_blobs</b> dataset with <b>4 centres</b> from scikit-learn (as in the lecture). Keep <code>y</code> only to score the clustering afterwards (K-Means itself is unsupervised).</div>

```python
X, y = datasets.make_blobs(
    n_samples=500,
    centers=4,
    n_features=2,
    shuffle=True,
    random_state=42
)
print("X shape:", X.shape)
print("y shape:", y.shape)
```

<div style="background:#16A34A;color:#FFFFFF;padding:12px 18px;border-radius:10px;margin:10px 0 14px 0;display:flex;align-items:center;gap:14px;box-shadow:0 2px 6px rgba(0,0,0,0.15);"><div style="background:rgba(255,255,255,0.22);color:#FFFFFF;font-weight:800;padding:6px 16px;border-radius:20px;font-size:22px;">3</div><div style="font-size:30px;font-weight:700;color:#FFFFFF;">Build the class</div></div>

<div style="font-size:20px;line-height:1.75;margin:6px 0;">Build a from-scratch <code>KMeans</code> class with a default <code>k</code>: initialise centroids from random points, then repeat <b>assign to nearest centroid</b> &rarr; <b>recompute centroids</b> until stable. <code>predict</code> returns each point's cluster id.</div>



$$\large \text{assign:}\quad c_i=\arg\min_{j}\;\lVert x_i-\mu_j\rVert$$

$$\large \text{update:}\quad \mu_j=\text{mean of the points assigned to cluster } j$$



```python
def euclidean_distance(x1, x2):
    return np.sqrt(np.sum((x1 - x2) ** 2))


class KMeans:
    def __init__(self, k=4, n_iteration=150):
        self.k = k
        self.n_iteration = n_iteration
        self.clusters = [[] for _ in range(self.k)]
        self.centroids = []

    def predict(self, X):
        self.X = X
        self.n_samples, self.n_features = X.shape

        # 1- Initialise centroids
        random_index = np.random.choice(
            self.n_samples,
            self.k,
            replace=False
        )

        self.centroids = [
            self.X[idx] for idx in random_index
        ]

        # 2- Optimisation loop
        for _ in range(self.n_iteration):

            # 2.1 Assign points to nearest centroid
            self.clusters = self._make_clusters(self.centroids)

            # 2.2 Recompute centroids
            old_centroids = self.centroids
            self.centroids = self._make_new_centroids(self.clusters)

            # 2.3 Check convergence
            if self._is_converged(old_centroids, self.centroids):
                break

        # 3- Return cluster label for every sample
        return self._get_cluster_labels(self.clusters)

    def _make_clusters(self, centroids):
        clusters = [[] for _ in range(self.k)]

        for idx, x in enumerate(self.X):
            centroid_idx = self._nearest_centroid(x, centroids)
            clusters[centroid_idx].append(idx)

        return clusters

    def _nearest_centroid(self, x, centroids):
        distances = [
            euclidean_distance(x, c)
            for c in centroids
        ]

        return np.argmin(distances)

    def _make_new_centroids(self, clusters):
        centroids = np.zeros((self.k, self.n_features))

        for idx, cluster in enumerate(clusters):
            centroids[idx] = np.mean(
                self.X[cluster],
                axis=0
            )

        return centroids

    def _is_converged(self, old_centroids, centroids):
        distances = [
            euclidean_distance(
                old_centroids[i],
                centroids[i]
            )
            for i in range(self.k)
        ]

        return sum(distances) == 0

    def _get_cluster_labels(self, clusters):
        labels = np.empty(self.n_samples)

        for cluster_idx, cluster in enumerate(clusters):
            for sample_idx in cluster:
                labels[sample_idx] = cluster_idx

        return labels
```

<div style="background:#7C3AED;color:#FFFFFF;padding:12px 18px;border-radius:10px;margin:10px 0 14px 0;display:flex;align-items:center;gap:14px;box-shadow:0 2px 6px rgba(0,0,0,0.15);"><div style="background:rgba(255,255,255,0.22);color:#FFFFFF;font-weight:800;padding:6px 16px;border-radius:20px;font-size:22px;">4</div><div style="font-size:30px;font-weight:700;color:#FFFFFF;">Train, predict &amp; report the metric</div></div>

<div style="font-size:20px;line-height:1.75;margin:6px 0;">K-Means is unsupervised, so score it with <b>clustering accuracy</b>: map each cluster to its <b>majority true label</b>, then measure how many points match.</div>

```python
# Create the model and predict the cluster ID for every point
model = KMeans(k=4, n_iteration=150)
y_predicted = model.predict(X).astype(int)

# Map each cluster to the majority true label
mapped_predictions = np.empty_like(y)

for cluster_id in range(4):
    cluster_mask = (y_predicted == cluster_id)

    if np.any(cluster_mask):
        true_labels_in_cluster = y[cluster_mask]
        majority_label = np.bincount(true_labels_in_cluster).argmax()
        mapped_predictions[cluster_mask] = majority_label

# Calculate clustering accuracy
accuracy = np.mean(mapped_predictions == y)

print("First 20 predicted cluster IDs:")
print(y_predicted[:20])

print("\nClustering Accuracy:", accuracy)
print("Clustering Accuracy Percentage:", round(accuracy * 100, 2), "%")
```
